```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)
```

# Matplotlib Visualization Library


https://matplotlib.org/stable/users/index.html

```python

fig, ax = plt.subplots()             # Create a figure containing a single Axes.
x = np.random.rand(5)
ax.plot([1, 2, 3, 4, 5], x)  # Plot some data on the Axes.
plt.show() 
```

Examples of simple plot 
https://matplotlib.org/stable/users/explain/quick_start.html#a-simple-example


```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.

ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.

ax.legend()  # Add a legend.
```

# Bar Plot
https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html

```python

# Generate random float numbers
random_floats = np.random.rand(5)  # Generate a random array of size 5


# Define the bar positions
bar_positions = np.arange(len(random_floats))

# Labels 
bar_labels = ['One', 'Two', 'Three', 'Four', 'Five']

# Plot the sorted floats as a bar plot
plt.bar(bar_positions,random_floats  , color='green')


# Add labels and title
plt.xticks(bar_positions, labels= bar_labels)
plt.xlabel('Random Floats')
plt.ylabel('Values')
plt.title('Bar Plot of Random Floating-Point Numbers')

# Add values on top of each bar
for i, (index, value) in enumerate(enumerate(random_floats)):
    plt.text(x=index, y=value , s=f'{value:.2f}', ha='center', va='bottom')


# Show the plot
plt.show()
```
